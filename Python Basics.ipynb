```python
'''
7. What is the difference between “==” and “is” operators in Python

== (Equality Operator) :
The == operator checks if the values of two objects are the same. It compares the contents of the objects.
It returns True if the values of the objects are equal, and False otherwise.
The == operator calls the __eq__ method (which is responsible for comparison) of the objects being compared.

Example of == operator -
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b)  # Output: True, because their values are equal (the lists contain the same elements)

is (Identity Operator) :
The is operator checks if two objects refer to the same memory location (i.e., if they are the same object in memory). It checks for object identity rather than equality of values.
It returns True if both objects point to the same object in memory (they are the exact same object), and False if they are two distinct objects, even if their contents are identical.
The is operator compares the memory address of the objects.

Example of is operator -
a = [1, 2, 3]
b = [1, 2, 3]
print(a is b)  # Output: False, because `a` and `b` are two different objects in memory, even though they have the same values.


== checks if the values of two objects are equal.

is checks if two objects are the same object in memory (i.e., if they refer to the same place in memory).
```

```python
'''
8. What are logical operators in Python

In Python, logical operators are used to combine conditional statements and return a Boolean value (True or False).
It allows to make more complex decisions in the code by combining multiple conditions.

There are three main logical operators in Python:

(i) and (Logical AND) -
The and operator returns True if both conditions are true.
If either condition is False, it returns False.
Example :
x = 10
y = 5
result = (x > 5) and (y < 10)  # Both conditions must be True
print(result)  # Output: True (because 10 > 5 and 5 < 10)
In the example above, the condition (x > 5) is True and (y < 10) is also True, so the result of and is True.

(ii) or (Logical OR) -
The or operator returns True if at least one condition is true.
It only returns False if both conditions are False.
Example :
x = 10
y = 5
result = (x < 5) or (y < 10)  # Only one condition needs to be True
print(result)  # Output: True (because 5 < 10 is True, even though 10 < 5 is False)
In this example, (x < 5) is False, but (y < 10) is True, so the result of or is True.

(iii) not (Logical NOT) -
The not operator is used to invert the truth value of a condition. It returns True if the condition is False, and False if the condition is True.
Example :
x = 10
result = not (x > 5)  # Inverts the result of (x > 5)
print(result)  # Output: False (because (x > 5) is True, and not True is False)
In this example, (x > 5) is True, so not (x > 5) returns False.


and: Returns True if both conditions are true.
or: Returns True if at least one condition is true.
not: Reverses the truth value of a condition.

 Above logical operators are commonly used in if statements and loops to control the flow of a program based on multiple conditions.
```

```python
 '''
 9. What is type casting in Python?

 Type casting in Python refers to the process of converting a variable from one data type to another.
 This is often necessary when you need to perform operations that require specific types of data, like converting a string to a number for calculations or converting an integer to a string for printing.

 Types of Type Casting in Python:

1. Implicit Type Casting (Type Coercion):
Python automatically converts a smaller data type (like int) to a larger data type (like float) when necessary. This is known as implicit type casting or type coercion.
For example, if you add an integer to a float, Python automatically converts the integer to a float to perform the operation.
Example -
x = 5        # int
y = 3.2      # float
result = x + y  # Python automatically converts x to float
print(result)   # Output: 8.2 (float)

2. Explicit Type Casting (Type Conversion):
You can manually convert one data type to another using Python’s built-in functions. This is known as explicit type casting.
For example, if you have a float but you need an integer, you can use int() to convert it. Or if you have an integer and you need a string, you can use str().
Common type casting functions:
int(): Converts a value to an integer.
float(): Converts a value to a float.
str(): Converts a value to a string.
list(): Converts an iterable (like a tuple or set) to a list.
tuple(): Converts an iterable to a tuple.
set(): Converts an iterable to a set.

Examples of Explicit Type Casting:

(i) Converting from float to int:
x = 10.7
y = int(x)  # Explicitly converting float to int (will truncate the decimal)
print(y)  # Output: 10

(ii) Converting from int to float:
a = 3
b = float(a)  # Converting int to float
print(b)  # Output: 3.0

(iii) Converting from int to string:
a = 42
b = str(a)  # Converting int to string
print(b)  # Output: "42"

(iv) Converting a string to a number:
str_num = "10"
num = int(str_num)  # Converting string to integer
print(num)  # Output: 10 (as integer)

Type casting is essential for ensuring that data types are compatible with operations or when you need to change the format of data in your program.
```

```python
'''
11. What is the purpose of conditional statements in Python?

The purpose of conditional statements in Python is to allow a program to make decisions based on certain conditions.
They enable the program to execute specific blocks of code depending on whether a condition is true or false.
Conditional statements let control the flow of a program by choosing which actions to take based on conditions.
They make program dynamic, allowing it to respond to different inputs or situations during execution.

Common Conditional Statements in Python :

(i) if statement -
The if statement is used to test a condition. If the condition evaluates to True, the associated block of code runs.
Example :
x = 10
if x > 5:
    print("x is greater than 5")
# Output: "x is greater than 5"
In this case, the condition x > 5 is True, so the print() statement inside the if block is executed.

(ii) else statement -
The else statement is used when the if condition is False. It provides an alternative block of code that will run if the condition isn't met.
Example :
x = 3
if x > 5:
    print("x is greater than 5")
else:
    print("x is less than or equal to 5")
# Output: "x is less than or equal to 5"

(iii) elif statement (short for "else if") -
The elif statement allows you to check multiple conditions. It’s used when you have more than two options to test. If the first if condition is False, the program checks the condition in elif. If that condition is True, the associated block of code runs.
You can use as many elif statements as needed.
Example :
x = 7
if x > 10:
    print("x is greater than 10")
elif x == 7:
    print("x is equal to 7")
else:
    print("x is less than 7")
# Output: "x is equal to 7"

(iv) Nested conditionals -
You can also nest conditional statements inside other conditional statements. This allows for more complex decision-making.
Example :
x = 10
if x > 5:
    if x % 2 == 0:
        print("x is greater than 5 and is even")
    else:
        print("x is greater than 5 and is odd")
else:
    print("x is less than or equal to 5")
# Output: "x is greater than 5 and is even"

By using conditionals effectively, we can create programs that react to different inputs, states, and situations, making the code more dynamic and flexible.
```

```python
'''
12. How does the elif statement work?

The elif statement in Python stands for "else if" and is used in conditional statements to test multiple conditions.
It's a way to check multiple expressions for True values and execute different blocks of code depending on which condition is True.
The elif statement is used after an if statement and before an optional else statement.

Syntax :
if condition1:
    # code block for condition1
elif condition2:
    # code block for condition2
elif condition3:
    # code block for condition3
else:
    # code block if none of the above conditions are True

Example :
x = 15

if x > 20:
    print("x is greater than 20")
elif x == 15:
    print("x is equal to 15")
elif x > 10:
    print("x is greater than 10 but less than 20")
else:
    print("x is 10 or less")

The program first checks if x > 20. Since x is 15, this condition is False, so it moves to the next condition.
The program then checks if x == 15. This condition is True, so the program prints "x is equal to 15" and skips the other conditions (including the else block).
If x were, for example, 12, it would have skipped the if and the first elif, and then checked the next elif condition (x > 10), which would have been True, printing "x is greater than 10 but less than 20".

> elif is used to check multiple conditions after an if statement.
> The program evaluates each condition in sequence, and once it finds a True condition, it executes that block of code and skips the rest.
> It's an efficient way to handle multiple possible conditions in a program.
```

```python
'''
13. What is the difference between for and while loops?

The main difference between for loops and while loops in Python lies in how they control the flow of execution and how they are used for iteration.

(1) for loop :
A for loop is used to iterate over a sequence (such as a list, tuple, string, or range).
It is typically used when the number of iterations is known or can be easily determined before the loop starts.

It is most commonly used when you need to iterate over a collection of items or a range of values.
The loop executes once for each element in the sequence.
The loop variable takes the value of each element in the sequence, one by one, during each iteration.

Syntax -
for variable in sequence:
    # code to be executed

Example -
for i in range(5):  # range(5) generates numbers from 0 to 4
    print(i)

In this example, the for loop iterates through the values from 0 to 4 (generated by range(5)) and prints each value.

(2) while loop :
A while loop repeatedly executes a block of code as long as a condition is True. The loop keeps executing until the condition evaluates to False.

The number of iterations is not fixed ahead of time; it depends on the condition you specify.
It is typically used when the number of iterations is not known in advance and you want the loop to run while a certain condition is true.

Syntax -
while condition:
    # code to be executed

Example -
i = 0
while i < 5:
    print(i)
    i += 1  # Increment the value of i

In this example, the while loop continues to execute as long as i < 5 is True.
The value of i is incremented with each iteration, and when i becomes 5, the condition i < 5 becomes False, and the loop stops.

> Example - Using a for loop to print the first 5 numbers :
for i in range(5):
    print(i)

> Example - Using a while loop to print the first 5 numbers :
i = 0
while i < 5:
    print(i)
    i += 1

for loop: Ideal when the number of iterations is known, or when iterating through a sequence of items.
It automatically handles the iteration and is more concise when iterating over a collection.

while loop: Ideal when the number of iterations is not known in advance and depends on a condition being met.
We must manually manage the loop variable and condition.
```

```python
#9. Write a Python program to convert user input from string to integer, float, and boolean types.

user_input = input("Enter a string value : ")

# convert the string input to integer
try :
  user_input_int = int(user_input)
  print(f"Converted string to integer is {user_input_int}")
except ValueError :
  print("input string cannot be converted into integer")

# convert the string input to float
try :
  user_input_float = float(user_input)
  print(f"Converted string to float is {user_input_float}")
except ValueError :
  print("inout string cannot be converted into float")

# convert the string input to boolean

if user_input.lower() in ['true', '1', 't', 'y', 'yes'] :
  user_input_bool = True
elif user_input.lower() in ['false', '0', 'f', 'n', 'no'] :
  user_input_bool = False
else :
  user_input_bool = bool(user_input)
print(f"Converted string to boolean is {user_input_bool}")
```

```python
#14. Write a program to reverse a string using a while loop.

string = input("Enter a string : ")
reversed_string = ""
index = len(string) - 1

while index >= 0 :
  reversed_string = reversed_string + string[index]
  index = index - 1

print("Reversed String : ", reversed_string)
```

```python
#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

num = int (input("Enter a number : "))

factorial = 1
i = 1

while i <= num :
  factorial = factorial * i

  i = i + 1

print(f"factorial of {num} = {factorial}")
```
